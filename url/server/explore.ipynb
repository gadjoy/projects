```javascript
const express = require('express');
const app = express();
const port = 5000;

// In-memory storage for URLs (use a database in production)
const urlDatabase = {};

// Function to generate a short URL
function shortenUrl(longUrl) {
    const shortCode = (Math.random() + 1).toString(36).substring(7); // Generate a random short code
    const shortUrl = `http://localhost:${port}/${shortCode}`;
    urlDatabase[shortCode] = longUrl; // Store mapping in memory
    return { shortUrl, shortCode };
}

// Endpoint to create a short URL
app.get('/shorten', (req, res) => {
    const longUrl = req.query.url;
    if (!longUrl) {
        return res.status(400).json({ error: 'URL parameter is required' });
    }
    const { shortUrl, shortCode } = shortenUrl(longUrl);
    res.json({ shortUrl, shortCode });
});

// Endpoint to redirect to the original URL
app.get('/:shortCode', (req, res) => {
    const shortCode = req.params.shortCode;
    const longUrl = urlDatabase[shortCode];

    if (longUrl) {
        res.redirect(longUrl);
    } else {
        res.status(404).send('Short URL not found');
    }
});

// Start the server
app.listen(port, () => {
    console.log(`URL shortener running at http://localhost:${port}`);
});
```

```javascript
// Import necessary modules
const http = require('http');
const url = require('url');

// In-memory database to store URL mappings
const urlDatabase = {};

// Function to generate a short code for a given URL
function generateShortCode(longUrl) {
    const shortCode = Math.random().toString(36).substring(2, 8);
    urlDatabase[shortCode] = longUrl;
    return shortCode;
}

// HTTP server to handle requests
const server = http.createServer((req, res) => {
    const parsedUrl = url.parse(req.url, true);
    
    // Endpoint to create a short URL
    if (parsedUrl.pathname === '/shorten') {
        const longUrl = parsedUrl.query.url;
        if (longUrl) {
            const shortCode = generateShortCode(longUrl);
            const shortUrl = `http://localhost:5000/${shortCode}`;
            
            // Respond with the shortened URL
            res.writeHead(200, { 'Content-Type': 'application/json' });
            res.end(JSON.stringify({ longUrl, shortUrl }));
        } else {
            res.writeHead(400, { 'Content-Type': 'text/plain' });
            res.end('URL parameter missing');
        }
    } 
    
    // Endpoint to handle redirection
    else {
        const shortCode = parsedUrl.pathname.slice(1); // Remove the leading "/"
        const longUrl = urlDatabase[shortCode];
        
        if (longUrl) {
            res.writeHead(302, { 'Location': longUrl });
            res.end();
        } else {
            res.writeHead(404, { 'Content-Type': 'text/plain' });
            res.end('Short URL not found');
        }
    }
});

// Start the server
server.listen(5000, () => {
    console.log('Server is running at http://localhost:5000');
});
```
