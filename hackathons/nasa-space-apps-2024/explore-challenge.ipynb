```python
import csv
import json

with open('challenges-2024.json', 'r') as file:
    data = json.load(file)

def get_challenge_info(challenge_data):
    title = challenge_data['node']['title']
    url = f"https://www.spaceappschallenge.org/nasa-space-apps-2024/challenges/{challenge_data['node']['meta']['relativeUrl']}"
    categories = [category['name'] for category in challenge_data['node']['categories']]
    skills = challenge_data['node']['skills']
    excerpt = challenge_data['node']['excerpt']
    return title, url, categories, skills, excerpt

# Compute the list of categories once
all_categories = [category['name'] for edge in data[0]['data']['challenges']['edges'] for category in edge['node']['categories']]

# Open a CSV file for writing
with open('challenges.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['Title', 'URL', 'Categories', 'Skills', 'Excerpt']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    # Write the header row
    writer.writeheader()

    # Iterate over the challenge edges and write data to the CSV file
    for edges in range(0, 21):
        challenge_edge = data[0]['data']['challenges']['edges'][edges]
        title, url, categories, skills, excerpt = get_challenge_info(challenge_edge)

        # Convert categories and skills to strings for writing to CSV
        categories_str = ', '.join(categories)
        skills_str = ', '.join(skills)

        # Write a row to the CSV file
        writer.writerow({'Title': title, 'URL': url, 'Categories': categories_str, 'Skills': skills_str, 'Excerpt': excerpt})
```
