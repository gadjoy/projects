```python
import random
import matplotlib.pyplot as plt

# Function to generate a shape without color
def generate_shape_without_color():
    shapes = ['circle', 'square', 'triangle']
    shape = random.choice(shapes)
    return shape

# Function to draw the shape without color
def draw_shape_without_color(shape):
    plt.figure()
    if shape == 'circle':
        circle = plt.Circle((0.5, 0.5), 0.4, edgecolor='black', facecolor='none')
        plt.gca().add_patch(circle)
    elif shape == 'square':
        square = plt.Rectangle((0.1, 0.1), 0.8, 0.8, edgecolor='black', facecolor='none')
        plt.gca().add_patch(square)
    elif shape == 'triangle':
        triangle = plt.Polygon([[0.5, 0.9], [0.1, 0.1], [0.9, 0.1]], edgecolor='black', facecolor='none')
        plt.gca().add_patch(triangle)
    plt.axis('scaled')
    plt.show()

# Generate and display the shape without color
shape = generate_shape_without_color()
draw_shape_without_color(shape)
```

```python
# Function to generate a color for the shape
def generate_color():
    colors = ['red', 'blue', 'green']
    color = random.choice(colors)
    return color

# Function to draw the shape with the generated color
def draw_shape_with_color(shape, color):
    plt.figure()
    if shape == 'circle':
        circle = plt.Circle((0.5, 0.5), 0.4, color=color)
        plt.gca().add_patch(circle)
    elif shape == 'square':
        square = plt.Rectangle((0.1, 0.1), 0.8, 0.8, color=color)
        plt.gca().add_patch(square)
    elif shape == 'triangle':
        triangle = plt.Polygon([[0.5, 0.9], [0.1, 0.1], [0.9, 0.1]], color=color)
        plt.gca().add_patch(triangle)
    plt.axis('scaled')
    plt.show()

# Generate color and apply it to the previously generated shape
color = generate_color()
draw_shape_with_color(shape, color)
```

```python
import random
import matplotlib.pyplot as plt

# Function to generate multiple shapes
def generate_multiple_shapes(num_shapes=3):
    three_shapes = ['circle', 'square', 'triangle']
    return [random.choice(three_shapes) for _ in range(num_shapes)]

# Function to generate multiple colors
def generate_multiple_colors(num_colors=3):
    colors = ['red', 'blue', 'green', 'yellow', 'purple']
    return [random.choice(colors) for _ in range(num_colors)]

# Function to draw multiple shapes with different colors
def draw_multiple_shapes_with_colors(shapes, colors):
    plt.figure()
    for i, (shape, color) in enumerate(zip(shapes, colors)):
        if shape == 'circle':
            circle = plt.Circle((0.5 + i*0.5, 0.5), 0.4, color=color)
            plt.gca().add_patch(circle)
        elif shape == 'square':
            square = plt.Rectangle((0.1 + i*0.5, 0.1), 0.4, 0.4, color=color)
            plt.gca().add_patch(square)
        elif shape == 'triangle':
            triangle = plt.Polygon([[0.5 + i*0.5, 0.9], [0.1 + i*0.5, 0.1], [0.9 + i*0.5, 0.1]], color=color)
            plt.gca().add_patch(triangle)
    plt.axis('scaled')
    plt.show()

# Generate shapes and colors
three_shapes_filled = generate_multiple_shapes(num_shapes=3)
colors = generate_multiple_colors(num_colors=3)

# Draw shapes with colors
draw_multiple_shapes_with_colors(three_shapes_filled, colors)

# Store shapes and colors for future reference
three_shapes_filled, colors
```
