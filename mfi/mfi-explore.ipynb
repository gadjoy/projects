```python
import requests
import pandas as pd

# Mutual funds and units data
mf = [
    {"id": "122639", "name": "Parag Parikh Flexi Cap Fund - Direct Plan - Growth"},
    {"id": "118778", "name": "Nippon India Small Cap Fund - Direct Plan - Growth"},
    {"id": "120251", "name": "ICICI Prudential Equity & Debt Fund - Direct Plan - Growth"},
]

units = [
    {"id": "122639", "units": 49.485},
    {"id": "118778", "units": 10.893},
    {"id": "120251", "units": 10.199},
]

base_url = "https://api.mfapi.in/mf"

def fetch_latest_nav(mf_id):
    url = f"{base_url}/{mf_id}/latest"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        return float(data['data'][0]['nav'])
    else:
        print(f"Failed to retrieve data for MF ID {mf_id}: {response.status_code}")
        return None

# List to store results
results = []

# Fetch NAV and calculate value for each fund
for fund in mf:
    mf_id = fund['id']
    nav = fetch_latest_nav(mf_id)
    if nav is not None:
        unit = next(item['units'] for item in units if item['id'] == mf_id)
        value = nav * unit
        results.append({
            "Fund Name": fund['name'],
            "Units": unit,
            "NAV": nav,
            "Value": value
        })

# Create DataFrame
df = pd.DataFrame(results)

# Calculate total net worth
total_net_worth = df['Value'].sum()

# Display the table and total net worth
print(df)
print(f"\nTotal Net Worth: {total_net_worth}")
```
