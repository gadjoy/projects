```python
import numpy as np
import pywt
import PIL.Image as Image

def embed_watermark(image_path, watermark_path, output_path):
    # Load images
    image = Image.open(image_path).convert("L")
    watermark = Image.open(watermark_path).convert("L")
    watermark = watermark.resize(image.size, Image.LANCZOS)

    # Convert images to numpy arrays
    image_np = np.array(image)
    watermark_np = np.array(watermark)

    # Apply Discrete Wavelet Transform (DWT)
    coeffs_image = pywt.dwt2(image_np, 'haar')
    cA, (cH, cV, cD) = coeffs_image

    # Resize watermark to match the size of cD
    watermark = watermark.resize(cD.shape[::-1], Image.LANCZOS)
    watermark_np = np.array(watermark)

    # Embed watermark in the diagonal component
    cD = cD + 0.1 * watermark_np  # Adjust the factor based on the visibility and robustness you need

    # Reconstruct the image with the watermark
    coeffs_watermarked = cA, (cH, cV, cD)
    watermarked_image = pywt.idwt2(coeffs_watermarked, 'haar')

    # Save watermarked image
    Image.fromarray(np.uint8(watermarked_image)).save(output_path)

def extract_watermark(image_path, watermark_path, extracted_path):
    # Load watermarked image
    watermarked_image = Image.open(image_path).convert("L")
    watermark = Image.open(watermark_path).convert("L")

    # Convert to numpy array
    watermarked_np = np.array(watermarked_image)
    watermark_np = np.array(watermark)

    # Apply Discrete Wavelet Transform (DWT)
    coeffs_watermarked = pywt.dwt2(watermarked_np, 'haar')
    _, (_, _, cD_watermarked) = coeffs_watermarked

    # Resize watermark to match the size of cD_watermarked
    watermark = watermark.resize(cD_watermarked.shape[::-1], Image.LANCZOS)
    watermark_np = np.array(watermark)

    # Extract watermark
    extracted_watermark = (cD_watermarked - 0.1 * watermark_np) / 0.1  # Adjust the factor based on the embedding

    # Save extracted watermark
    Image.fromarray(np.uint8(extracted_watermark)).save(extracted_path)

# Example usage
embed_watermark('artist_image.jpeg', 'qr_code.jpg', 'watermarked_image.jpg')
extract_watermark('watermarked_image.jpg', 'qr_code.jpg', 'extracted_watermark.jpg')
```

```python
import PIL.Image as Image
import pywt
import numpy as np

def embed_watermark(image_path, watermark_path, output_path, strength=0.05):
    # Load images
    image = Image.open(image_path).convert("L")
    watermark = Image.open(watermark_path).convert("L")

    # Apply Discrete Wavelet Transform (DWT)
    coeffs_image = pywt.dwt2(np.array(image), 'haar')
    cA, (cH, cV, cD) = coeffs_image

    # Resize watermark to match the dimensions of the diagonal coefficients
    watermark = watermark.resize(cD.shape[::-1], Image.Resampling.LANCZOS)
    watermark_np = np.array(watermark)

    # Embed watermark in the diagonal component
    cD += strength * watermark_np  # Reduced strength for less distortion

    # Reconstruct the image with the watermark
    watermarked_image = pywt.idwt2((cA, (cH, cV, cD)), 'haar')

    # Save watermarked image
    Image.fromarray(np.uint8(watermarked_image.clip(0, 255))).save(output_path)

def extract_watermark(image_path, watermark_path, extracted_path, strength=0.05):
    # Load watermarked image and watermark template
    watermarked_image = Image.open(image_path).convert("L")
    watermark = Image.open(watermark_path).convert("L")

    # Apply Discrete Wavelet Transform (DWT)
    coeffs_watermarked = pywt.dwt2(np.array(watermarked_image), 'haar')
    _, (_, _, cD_watermarked) = coeffs_watermarked

    # Resize watermark to match the dimensions of the diagonal coefficients for extraction
    watermark = watermark.resize(cD_watermarked.shape[::-1], Image.Resampling.LANCZOS)
    watermark_np = np.array(watermark)

    # Extract watermark
    extracted_watermark = (cD_watermarked - strength * watermark_np) / strength

    # Save extracted watermark
    Image.fromarray(np.uint8(extracted_watermark.clip(0, 255))).save(extracted_path)

# Example usage
# embed_watermark('original_image.jpg', 'qr_code.jpg', 'watermarked_image.jpg', strength=0.02)
# extract_watermark('watermarked_image.jpg', 'qr_code.jpg', 'extracted_qr_code.jpg', strength=0.02)

embed_watermark('artist_image.jpeg', 'qr_code.png', 'watermarked_image.jpg', strength=0.02)
extract_watermark('watermarked_image.jpg', 'qr_code.png', 'extracted_watermark.jpg', strength=0.02)
```
